```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Function to calculate cost (mean squared error)
def cost_function(theta0, theta1, X, y):
    m = len(y)
    h = theta0 + theta1 * X
    return (1 / (2 * m)) * np.sum((h - y)**2)

# Function to perform gradient descent
def gradient_descent(X, y, theta0, theta1, learning_rate, iterations):
    m = len(y)
    costs = []

    for iteration in range(iterations):
        h = theta0 + theta1 * X
        delta_theta0 = (1 / m) * np.sum(h - y)
        delta_theta1 = (1 / m) * np.sum((h - y) * X)

        theta0 -= learning_rate * delta_theta0
        theta1 -= learning_rate * delta_theta1

        cost = cost_function(theta0, theta1, X, y)
        costs.append(cost)

    return theta0, theta1, costs

# Plotting the linear regression
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, y, label='Data Points')
theta0, theta1, _ = gradient_descent(X, y, 0, 0, learning_rate=0.01, iterations=1000)
plt.plot(X, theta0 + theta1 * X, color='red', label='Best Fit Line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Plotting the contour plot
plt.subplot(1, 2, 2)
theta1_values = np.linspace(0, 6, 100)
theta0_values = np.linspace(0, 8, 100)
T0, T1 = np.meshgrid(theta0_values, theta1_values)
J = np.array([cost_function(t0, t1, X, y) for t0, t1 in zip(np.ravel(T0), np.ravel(T1))])
J = J.reshape(T0.shape)
plt.contour(T0, T1, J, levels=np.logspace(0, 5, 20), cmap='viridis', alpha=0.7)
plt.colorbar(label='Cost Function (J)')
plt.scatter(theta0, theta1, color='red', marker='x', label='Optimal Parameters')
plt.title('Contour Plot')
plt.xlabel('Theta0')
plt.ylabel('Theta1')
plt.legend()

plt.tight_layout()
plt.show()
```

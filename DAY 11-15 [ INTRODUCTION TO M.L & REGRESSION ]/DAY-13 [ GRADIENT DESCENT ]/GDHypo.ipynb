```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('placement.csv')
df.head()

# Feature Scaling
def normalize_feature(feature):
    return (feature - np.mean(feature)) / np.std(feature)

x = normalize_feature(df.iloc[:, 0:1].values)  # Normalizing CGPA
y = df.iloc[:, -1].values

def hypothesis(theta0, theta1, x):
    return theta0 + theta1 * x

def cost_function(theta0, theta1, x, y):
    m = len(y)
    h = hypothesis(theta0, theta1, x)
    return (1 / (2 * m)) * np.sum((h - y)**2)

def gradient_descent(theta0, theta1, x, y, learning_rate=0.0001, iterations=1000, convergence_threshold=1e-4):
    m = len(y)
    costs = []  # To store cost values for each iteration
    for iteration in range(iterations):
        h = hypothesis(theta0, theta1, x)
        delta_theta0 = (1 / m) * np.sum(h - y)
        delta_theta1 = (1 / m) * np.sum((h - y) * x)
        
        theta0 -= learning_rate * delta_theta0
        theta1 -= learning_rate * delta_theta1

        # Check for convergence based on gradients
        if np.abs(delta_theta0) < convergence_threshold and np.abs(delta_theta1) < convergence_threshold:
            print(f"Converged after {iteration} iterations.")
            break

        # Store cost for every iteration
        cost = cost_function(theta0, theta1, x, y)
        costs.append(cost)

        # Print cost for every 100 iterations
        if iteration % 100 == 0:
            print(f"Iteration {iteration}, Cost: {cost}")

    return theta0, theta1, costs

# Initialize theta0 and theta1
theta0 = 1.0
theta1 = 1.0

# Run gradient descent
final_theta0, final_theta1, costs = gradient_descent(theta0, theta1, x, y)

# Plotting the cost over iterations
iterations_range = range(0, len(costs) * 100, 100)
plt.plot(iterations_range, costs, marker='o')
plt.title('Cost over Iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Print the final parameters
print(f"Final Parameters: theta0 = {final_theta0}, theta1 = {final_theta1}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data from CSV file
# Make sure to replace 'placement.csv' with the actual file path
data = pd.read_csv('placement.csv')

# Extract features (X) and target variable (y) from the data
X = data.iloc[:,0:1].values
y = data.iloc[:, -1].values

# Feature Scaling for gradient descent
def normalize_feature(feature):
    return (feature - np.mean(feature)) / np.std(feature)

x_normalized = normalize_feature(X)

# Gradient Descent
def hypothesis(theta0, theta1, x):
    return theta0 + theta1 * x

def cost_function(theta0, theta1, x, y):
    m = len(y)
    h = hypothesis(theta0, theta1, x)
    return (1 / (2 * m)) * np.sum((h - y)**2)

def gradient_descent(theta0, theta1, x, y, learning_rate=0.0001, iterations=1000, convergence_threshold=1e-4):
    m = len(y)
    costs = []  # To store cost values for each iteration
    for iteration in range(iterations):
        h = hypothesis(theta0, theta1, x)
        delta_theta0 = (1 / m) * np.sum(h - y)
        delta_theta1 = (1 / m) * np.sum((h - y) * x)
        
        theta0 -= learning_rate * delta_theta0
        theta1 -= learning_rate * delta_theta1

        # Check for convergence based on gradients
        if np.abs(delta_theta0) < convergence_threshold and np.abs(delta_theta1) < convergence_threshold:
            print(f"Converged after {iteration} iterations.")
            break

        # Store cost for every iteration
        cost = cost_function(theta0, theta1, x, y)
        costs.append(cost)

        # Print cost for every 100 iterations
        if iteration % 100 == 0:
            print(f"Iteration {iteration}, Cost: {cost}")

    return theta0, theta1, costs

# Initialize theta0 and theta1 for gradient descent
theta0_gd = 1.0
theta1_gd = 1.0

# Run gradient descent
final_theta0_gd, final_theta1_gd, costs = gradient_descent(theta0_gd, theta1_gd, x_normalized, y)

# Plotting the cost over iterations for gradient descent
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
iterations_range = range(0, len(costs) * 100, 100)
plt.plot(iterations_range, costs, marker='o')
plt.title('Cost over Iterations (Gradient Descent)')
plt.xlabel('Iterations')
plt.ylabel('Cost')

# Linear Regression using sklearn
plt.subplot(1, 2, 2)
# Fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Get the slope (coef_) and intercept (intercept_) of the best-fit line
slope = model.coef_[0]
intercept = model.intercept_

# Plot the scatter plot and best-fit line for sklearn
plt.scatter(X, y, label='Training Data')
plt.plot(X, model.predict(X), color='red', label=f'Best Fit Line (sklearn): y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Feature')
plt.ylabel('Target Variable')
plt.title('Scatter Plot and Best Fit Line (sklearn)')
plt.legend()

plt.tight_layout()
plt.show()

# Print the final parameters for gradient descent
print(f"\nFinal Parameters (Gradient Descent): theta0 = {final_theta0_gd}, theta1 = {final_theta1_gd}")
```

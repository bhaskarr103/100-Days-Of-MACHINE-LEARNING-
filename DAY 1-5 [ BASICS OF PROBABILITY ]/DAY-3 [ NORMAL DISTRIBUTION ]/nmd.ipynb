# NORMAL DISTRIBUTION

```python
import seaborn as sns

x = sns.load_dataset('titanic')

x.head()
```

```python
sns.kdeplot(x['age']) # Normal Distribution
```

```python
x['age'].mean() # ==>  29.69
```

```python
x['age'].std()  # ==> 14.52
```

**N(29.69,14.52) ----> N(0,1) (Changing to Standard Variate)**

**Z = (X - μ) / σ**

```python
y = (x['age']-x['age'].mean())/x['age'].std()
y.head()
```

```python
sns.kdeplot(y)
```

# VISUALIZING NORMAL DISTRIBUTION

Normal distribution fits the probability distriutuion of many events.
Use **random.normal()** method to get a normal data distribution.

**loc -**(mean) where the beak of bell exists.
**scale -**(std.dev.) how flat the graph distribution should be
**size -**the shape of returned array

```python
from numpy import random
import seaborn as sns

x = random.normal(size=(2,3))
x
# generate a random normal dist of size(2 X 3).
```

```python
# set mean and std dev of dist

x = random.normal(loc=1,scale=2,size=(2,3))
x
```

```python
import warnings
warnings.filterwarnings('ignore')

sns.distplot(x)
```

```python
import warnings
warnings.filterwarnings('ignore')

sns.distplot(x,hist=False)
```

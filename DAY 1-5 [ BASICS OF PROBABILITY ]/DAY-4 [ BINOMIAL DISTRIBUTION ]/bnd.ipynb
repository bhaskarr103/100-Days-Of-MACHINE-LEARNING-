# BINOMIAL DISTRIBUTION

**i.e,**
**[ tossing 10 coin ] ----> performing this event 1000 times**

n = nos of trials
p = probability of success
size = nos of sample to generate

```python
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

n = 10 # tossing 10 coin
p = 0.5 # prob of success occuring
size = 1000 # performing event 1000 times.

binomial_dist = np.random.binomial(n,p,size)

sns.distplot(binomial_dist, hist=True)
```

**p tends to 1, histogram shift toward right**
